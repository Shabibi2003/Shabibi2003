```python
columns = df[['age','campaign','duration']]
Q1 = np.percentile(columns,25)  #25% of columns
Q3 = np.percentile(columns,75)

# IQR = Inter Quartile Range {Q3-Q1}
iqr = Q3-Q1

lower_limit = Q1 - 1.5 * iqr
upper_limit = Q3 + 1.5 * iqr

df[['age','campaign','duration']] = columns[(columns > lower_limit) & (columns < upper_limit)]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Lb = LabelEncoder()  # used to convert the categarical labels into numeric values
```

```python
X = df_encoded.drop('y',axis=1)  # X feature (independent variable)
```

```python
y = df_encoded['y']   # dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
def eval_model(y_test,y_prediction):
    accuracy = accuracy_score(y_test,y_prediction)
    print("Accuracy :",accuracy)
    
    cm = confusion_matrix(y_test,y_prediction)
    print("Confusion Matrix\n",cm)
    
    cl = classification_report(y_test,y_prediction)
    print("Classification Report: \n",cl)
    
def score(model):
    train_score = model.score(X_train,y_train)
    test_score = model.score(X_test,y_test)
    
    print("Training Score",train_score)
    print("Test Score",test_score)
    
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(X_train,y_train)
```

```python
score(dt)
```

```python
y_pred_dt = dt.predict(X_test)
```

```python
eval_model(y_test,y_pred_dt)
```
